```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r"C:\Users\N LAKSHMI PRASANNA\OneDrive\Attachments\Desktop\health_care\healthcare-dataset-stroke-data .csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
sns.boxplot(x='stroke', y='bmi', data=df)
plt.title("BMI Distribution by Stroke Status")
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='age', y='avg_glucose_level', hue='stroke', palette='Set1')
plt.title('Age vs Glucose Level by Stroke Status')
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='work_type', hue='stroke', palette='viridis')
plt.title('Stroke Count by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Stroke')
plt.tight_layout()
plt.show()
```

```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#FF9999','#66B3FF','#99FF99'])
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio to make it a circle
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='work_type', y='avg_glucose_level', palette='Blues_d')
plt.title('Average Glucose Level by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Avg Glucose Level')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by age and calculate average glucose
glucose_by_age = df.groupby('age')['avg_glucose_level'].mean().reset_index()

# Plot the line chart
plt.figure(figsize=(10,5))
sns.lineplot(data=glucose_by_age, x='age', y='avg_glucose_level', color='blue')
plt.title('Average Glucose Level by Age')
plt.xlabel('Age')
plt.ylabel('Avg Glucose Level')
plt.grid(True)
plt.show()
```
